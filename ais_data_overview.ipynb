```python
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
ais_data = pd.read_csv('../data/ais/aisdk_20171101.csv')
```

```python
longitudes = ais_data.Longitude.values
```

```python
plt.figure(figsize=(16,7))
plt.hist(longitudes, bins=1000)
plt.xlim([-180, 180])
plt.xlabel('Longitude, degrees')
plt.ylabel('Longitude, counts')
plt.title('Histogram of Longitude Distribution, November 1')
plt.show()
```
